```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = df.drop("Unnamed: 0",axis =1)
print(df.head())
```

Gender Distribution

```python
plt.figure(figsize = (5,5))
custom_colors = ["#3498db", "#e74c3c"] 
ax = sns.countplot(data=df, x="Gender", hue="Gender", palette=custom_colors, dodge=False)
#ax.bar_label(ax.containers[0])
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

#From the above chart we have analysed that:
#the number of females in the data is more than the number of males

```python
gp = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gp)
```

```python
#heatmap

plt.figure(figsize = (4,4))
sns.heatmap(gp, annot= True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

From the above chart we have concluded that:
the education of the parents have a good impact on their scores

```python
gp1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gp1)
```

```python
#heatmap

plt.figure(figsize = (4,4))
sns.heatmap(gp1, annot= True) #write values inside colors
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

From the above chart we have concluded that:
There is negligible impact on the student's score due to their parent's marital status

Distribution of Ethnic Groups

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

mlist = [groupA["EthnicGroup"],
         groupB["EthnicGroup"],
        groupC["EthnicGroup"],
        groupD["EthnicGroup"],
        groupE["EthnicGroup"]]


data = {
    'EthnicGroup': ['group A', 'group B', 'group C', 'group A', 'group B', 'group D', 'group E', 'group C', 'group A', 'group E']
}
df = pd.DataFrame(data)

# Counting occurrences for each group
group_counts = df['EthnicGroup'].value_counts()

# Prepare the data for the pie chart
mlist = group_counts.values
labels = group_counts.index

# Define custom colors
custom_colors = ["#3498db", "#e74c3c", "#f1c40f", "#2ecc71", "#9b59b6"]

# Create the pie chart with enhancements
fig, ax = plt.subplots(figsize=(10, 7))
wedges, texts, autotexts = ax.pie(mlist, labels=labels, autopct="%1.2f%%", colors=custom_colors, shadow=True, startangle=140, explode=(0.1, 0, 0, 0, 0), textprops=dict(color="w"))

# Customizing texts and autotexts
for text in texts:
    text.set_fontsize(14)
    text.set_fontweight('bold')

for autotext in autotexts:
    autotext.set_fontsize(12)
    autotext.set_fontweight('bold')

# Adding a legend
ax.legend(wedges, labels, title="Ethnic Groups", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Title with a modern font
plt.title("Distribution of Ethnic Groups", fontsize=18, fontweight='bold')

# Adding annotations for better understanding
plt.annotate('Largest Group', xy=(0.1, 0.1), xytext=(1.2, 0.4),
             arrowprops=dict(facecolor='black', arrowstyle="->", lw=2),
             fontsize=12, fontweight='bold')

# Show plot
plt.show()
```
